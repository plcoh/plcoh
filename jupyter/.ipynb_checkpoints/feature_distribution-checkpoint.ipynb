```python
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
from ppo.audio_feature import AudioFeature
from services.feature_service import FeatureService
```

```python
mpl.rcParams['figure.dpi'] = 200
sns.set_style("whitegrid")
alpha=0.9
```

```python
feature_service = FeatureService()
feature_service.load_from_cache()
```

```python
feature_size = len(AudioFeature.feature_labels)
feature_size
```

```python
f, axis = plt.subplots(5, 2, figsize=(10, 14))
for i, label in enumerate(AudioFeature.feature_labels):
    x, y = i // 2, i % 2
    if i == 2:
        features = np.array([v[label] for v in feature_service.features.values()])
        print(max(features))
        ax = sns.histplot(features, ax=axis[x, y], bins=65, alpha=alpha)
        ax.set_xlim(-60, 5)
        ax.set_xlabel(label)
        if y > 0:
            ax.set_ylabel(None)
    elif i < 9:
        features = np.array([v[label] for v in feature_service.features.values()])
        ax = sns.histplot(features, ax=axis[x, y], bins=50, alpha=alpha)
        ax.set_xlabel(label)
        if y > 0:
            ax.set_ylabel(None)
    elif i == 9:
        features = np.array([v[label] for v in feature_service.features.values()])
        df = pd.DataFrame(features, columns=['mode', 'key'])
        ax = sns.countplot(df, x='key', hue='mode', ax=axis[x, y], alpha=alpha)
        ax.set_xticklabels(['c', 'c#', 'd', 'd#', 'e', 'f', 'f#', 'g', 'g#', 'a', 'bb', 'b'])
        ax.set_xlabel('mode/key (tonality)')
        plt.legend(title='mode', loc='upper right', labels=['minor(0)', 'major(1)'])
        if y > 0:
            ax.set_ylabel(None)
plt.tight_layout()
plt.show()
plt.savefig('plots/fig_feature_dist.png')
```
